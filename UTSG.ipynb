```python
Simulation_time=500
dt=.01

t=0
tempi=[]
temp1=[]
temp2=[]
temp3=[]
temp4=[]
temp0=[]
time=[]
logo()
i=0
UTSG.Wp1=UTSG.Win

while t<Simulation_time:


    Tpi=UTSG.integrator(UTSG.DTpi,argsforfunction=[330+273.016],intitial_cond=UTSG.Tpi,time_step=dt)
    Tpo=UTSG.integrator(UTSG.DTpo,argsforfunction=[],intitial_cond=UTSG.Tpo,time_step=dt)

    L_w=UTSG.integrator(UTSG.DLs1,argsforfunction=[],intitial_cond=UTSG.L_w,time_step=dt)
    
    Tp1=UTSG.integrator(UTSG.DTp1,argsforfunction=[],intitial_cond=UTSG.Tp1,time_step=dt)
    Tp2=UTSG.integrator(UTSG.DTp2,argsforfunction=[],intitial_cond=UTSG.Tp2,time_step=dt)
    Tp3=UTSG.integrator(UTSG.DTp3,argsforfunction=[],intitial_cond=UTSG.Tp3,time_step=dt)
    Tp4=UTSG.integrator(UTSG.DTp4,argsforfunction=[],intitial_cond=UTSG.Tp4,time_step=dt)

    Tm1=UTSG.integrator(UTSG.DTm1,argsforfunction=[],intitial_cond=UTSG.Tm1,time_step=dt)
    Tm2=UTSG.integrator(UTSG.DTm2,argsforfunction=[],intitial_cond=UTSG.Tm2,time_step=dt)
    Tm3=UTSG.integrator(UTSG.DTm3,argsforfunction=[],intitial_cond=UTSG.Tm3,time_step=dt)
    Tm4=UTSG.integrator(UTSG.DTm4,argsforfunction=[],intitial_cond=UTSG.Tm4,time_step=dt)

    Tsat=UTSG.integrator(UTSG.DTsat,argsforfunction=[],intitial_cond=UTSG.Tsat,time_step=dt)
    '''rho_b=UTSG.integrator(UTSG.Drho_b,argsforfunction=[],intitial_cond=UTSG.rho_b,time_step=dt)
    h_b=UTSG.integrator(UTSG.Dh_b,argsforfunction=[],intitial_cond=UTSG.hb,time_step=dt)
    rho_r=UTSG.integrator(UTSG.Drho_r,argsforfunction=[],intitial_cond=UTSG.rho_r,time_step=dt)
    Ldw=UTSG.integrator(UTSG.DLdw,argsforfunction=[],intitial_cond=UTSG.Ldw,time_step=dt)'''
    Td=UTSG.integrator(UTSG.DTd,argsforfunction=[],intitial_cond=UTSG.Td,time_step=dt)
    


    

    UTSG.Tpi=Tpi
    UTSG.Tpo=Tpo
    UTSG.Tp1=Tp1
    UTSG.Tp2=Tp2
    UTSG.Tp3=Tp3
    UTSG.Tp4=Tp4

    UTSG.Tm1=Tm1
    UTSG.Tm2=Tm2
    UTSG.Tm3=Tm3
    UTSG.Tm4=Tm4
    #print(Tm1+Tm2-UTSG.Ts1)
    
    UTSG.Tsat=Tsat
    #UTSG.Tdw=Tdw
    UTSG.Td=Td
    
    UTSG.L_w=L_w
    #UTSG.Ldw=Ldw

    #UTSG.rho_b=rho_b
    #UTSG.rho_g=rho_g
    #UTSG.hb=h_b

    if i%1000==0:

        time.append(t)
        tempi.append(Tpi)
        temp1.append(Tp1)
        temp2.append(Tp2)
        temp3.append(Tp3)
        temp4.append(Tp4)
        temp0.append(Tsat)
    
    #print('%.3f'%t,'', '%.3f'%(UTSG.Tpi-UTSG.Tp1),'%.3f'%(-UTSG.Tp1+UTSG.Tm1),'%.3f'%(UTSG.Tp1),'%.3f'%UTSG.Tpo,'%.3f'%(UTSG.Tp4-UTSG.Tpo))
    print('%.3f'%UTSG.DTsat())
    i=i+1


    
    t=t+dt
    
```
